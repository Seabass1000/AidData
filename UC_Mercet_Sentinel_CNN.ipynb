```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf
   
from tensorflow import keras
from tensorflow.keras import layers,models
from tensorflow.keras.models import Sequential

import os
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tqdm import tqdm
from tqdm.notebook import tqdm
import seaborn as sns


from sklearn.metrics import confusion_matrix
from sklearn import model_selection
from sklearn.model_selection import train_test_split
```

```python
# exploring data
import pathlib

data_dir = '/Users/test/Documents/Fall 2021/AID Data/UCMerced_LandUse/Images'

labels = ['agricultural','airplane','baseballdiamond','beach','buildings', 'chaparral','denseresidential',
         'forest','freeway','golfcourse','harbor','intersection','mediumresidential','mobilehomepark',
         'overpass','parkinglot','river','runway','sparseresidential','storagetanks','tenniscourt']

# load training data

img_size = 96

training_data = []

def load_training_data():
    for label in labels:
        path = os.path.join(data_dir, label)
        class_num = labels.index(label)
        for img in os.listdir(path):
            img_array = cv2.imread(os.path.join(path, img))
            # resize images
            new_array = cv2.resize(img_array, (img_size, img_size))
            training_data.append([new_array, class_num])
            
load_training_data()          

# shuffle data
import random
random.shuffle(training_data)
```

```python
print(type(training_data[0][0][0][0][0]))
# training data is a lsit of an array of an array of an integer
```

```python
# lists to store features and labels for training data
X = []
y = []


for features, label in training_data:
    X.append(features)
    y.append(label)
    
X = np.array(X).reshape(-1, img_size, img_size, 3)
y = np.array(y)
```

```python
model = models.Sequential()
model.add(layers.Rescaling(1./255, input_shape=(img_size, img_size, 3)))

model.add(layers.Conv2D(3, (7, 7), strides = (2, 2), activation = 'relu', use_bias=True, kernel_regularizer =tf.keras.regularizers.l2( l=0.01), input_shape=(96, 96, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation = 'relu', use_bias=True, kernel_regularizer =tf.keras.regularizers.l2( l=0.01)))
model.add(layers.Dropout(0.1))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(192, (3, 3), activation = 'relu', use_bias=True, kernel_regularizer =tf.keras.regularizers.l2( l=0.01)))
model.add(layers.Dropout(0.2))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(192, (2, 2), activation = 'relu', use_bias=True, kernel_regularizer =tf.keras.regularizers.l2( l=0.01)))
model.add(layers.Dropout(0.3))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.4))

model.add(layers.Flatten())
model.add(layers.Dense(64, activation = "relu"))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(64, activation = "relu"))
model.add(layers.Dense(21, activation = "softmax"))
```

```python
# Set a compiler

model.compile(optimizer='adam',
              metrics=['accuracy'])
```

```python
# Set a learning rate annealer
learning_rate_reduction = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', 
                                            patience=3, 
                                            verbose=1, 
                                            factor=0.5, 
                                            min_lr=0.005)
```

```python
# Fit the data
history = model.fit(X, y, epochs=1000, 
                    validation_split=0.2, 
                    callbacks=[learning_rate_reduction])
```
